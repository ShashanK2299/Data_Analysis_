# skewness and outlier 

```python
import seaborn as sns
sns.boxplot(data['Age'])
```

```python
sns.boxplot(data['Fare'])
```

```python
print('Skewness value of Age:', data['Age'].skew())
```

```python
print('Skewness value of Fare:', data['Fare'].skew())
```

```python
print('Skewness value of Survived:', data['Survived'].skew())
```

#### The skewness should be between the -1 to 1.

#### If we have any major changes from the value range that may indicate the presence of outliers

```python
median = float(data['Fare'].median())
```

```python
median
```

```python
import numpy as np
data['Fare']= np.where(data['Fare']> median, median, data['Fare'])
```

```python
sns.boxplot(data['Fare'])
```

```python
data_new= pd.DataFrame({'Colours':['Red','Yellow','Blue','Green']})
data_new
```

### create object of Ordinal Encoding

```python
encoder= ce.OrdinalEncoder(cols=['Colours'], return_df= True, mapping=[{'col':'Colours','mapping':{'None':0,'Red':1,'Yellow':2,'Blue':3,'Green':4}}])
```

```python
data_new_1=encoder.fit_transform(data_new)
```

```python
data_new_1
```

```python
data_new['encoder'] = encoder.fit_transform(data_new)
```
